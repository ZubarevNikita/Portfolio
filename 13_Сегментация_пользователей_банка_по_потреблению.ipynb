# Сегментация пользователей по потреблению

Нам, как аналитикам "Метанпромбанка" необходимо провести сегментацию пользователей по потреблению.  

Необходимо проанализировать клиенов регионального банка и сегментировать по количеству потребляемых продуктов, обращая внимание на отток.  

## Этапы выполнения проекта

1. Загрузка данных и подготовка к исследованию  
    1.1. Загрузка данных из источника, проверка их целостности;  
    1.2. Импорт необходимых библиотек;  
    1.3. Преобразование данных в удобный для анализа формат;

2. Подготовка данных  
    2.1. Проверка на дубликаты (явные и неявные);  
    2.2. Исследование и обработка аномалий;  
    2.3. Обработка пропусков или выявление причин их возникновения, выбор метода заполнения пропусков;    
    2.4. Кодирование категориальных признаков;  
    
3. Исследовательский анализ данных  
    3.1. Анализ средних значений признаков с разбивкой по отточности;  
    3.2. Изучение распределений признаков с разбивкой по отточности;  
    3.3. Изучение распределений признаков без привязки к признаку отточности;  
    3.4. Анализ корреляционный;  
    3.5. Формирование портрета клиента (или портретов, с разделением по признаку отточности)  
    3.6. Анализ портретов клиентов, сравнение между собой;  
4. Проверка и формирование статистических гипотез  
    4.1. Проверка гипотезы о различии дохода между теми клиентами, которые пользуются двумя продуктами банка, и теми, которые пользуются одним;  
    4.1.  Формулирование и проверка дополнительных гипотез;  
5. Определение показателей, которые стратегически влияют на потребление продуктов  
6. Сегментация на основе продуктов и стратегических показателей  
    6.1. Приоритезация сегментов  
7. Выводы и рекомендации для заказчика  
    7.1. Конкретные рекомендации для каждого сегмента клиентов, основанные на результате сегментации и анализе данных;  
    7.2. Рекомендации по продуктам, которые могут быть наиболее востребованы клиентами в каждом сегменте
8. Формирование отчета
    8.1. Подготовка дашборда;
    8.2. Подготовка презентации.

## Описание данных

Датасет содержит данные о клиентах банка «Метанпром». Банк располагается в Ярославле и областных городах: Ростов Великий и Рыбинск.  

Колонки:  
`USERID`— идентификатор пользователя,  
`score`— баллы кредитного скоринга,  
`city`— город,  
`gender`— пол,  
`age`— возраст,  
`equity`— приблизительная оценка собственности клиента,  
`balance` — баланс на счёте,  
`products`— количество продуктов, которыми пользуется клиент,  
`credit_card`— есть ли кредитная карта,  
`last_activity` — был ли клиент активен последнее время,  
`EST_SALARY`— заработная плата клиента,  
`churn`— уходит или нет.

## Изучение распределений признаков без привязки к отточности

Создадим функцию для построения kde-графиков с указанием экстремума.

```python
def viz_kde(data, x, hue=None, title='Введите название графика'):
    '''Строит kde-графики с указанием экстремума

       data-датасет
       x-изучаемая величина
       hue-по какому показателю разбивать график на цвета
       title-название графика
    '''

    viz = sns.kdeplot(data=data, x=x, hue=hue, common_norm=False)
    # Находим положение экстремума на оси x
    for i in range(len(data[hue].unique())):
        kde_max_idx = viz.get_lines()[i].get_data()[1].argmax()
        kde_max = viz.get_lines()[i].get_data()[0][kde_max_idx]

    # Добавляем вертикальную линию на месте экстремума
        plt.axvline(x=kde_max, color='gray', linestyle='--', alpha=0.5)
        plt.title(title)
        sns.set_style('white')
```

### Построим графики для небинарных признаков с разбиением по количеству продуктов

Возраст клиентов с разным количеством продуктов отличается незначительно.

#### Balance

## Корреляционный анализ

```python
clients.columns
```

Исследуем корреляцию между параметрами с помощью [Phik(𝜙k)-метода.](https://towardsdatascience.com/phik-k-get-familiar-with-the-latest-correlation-coefficient-9ba0032b37e7)  
Основное преимущество: может определять нелинейные зависимости.

```python
# создадим датасет с необходимыми нам столбцами
for_corr = clients[['score', 'age', 'equity', 'balance',
                    'products', 'credit_card', 'last_activity', 'est_salary', 'churn',]]

phik_corr = for_corr.phik_matrix()
phik_corr.round(2)
sns.heatmap(phik_corr, annot=True, fmt='0.2f',
            linewidths=.5, cmap="vlag", center=0)
plt.title('Матрица корреляции')
plt.show()
```

Выявленные корреляции:  
- Значение кредитного рейтинга (`Score`) коррелирует с оценкой собственности клиента(`equity`), количеством продуктов(`products`) и зарплатой клиента(`est_salary`);  
- Возраст клиента коррелирует с заработной платой;

# Проверка и формирование статистических гипотез

## Проверка гипотезы о различии дохода между теми клиентами, которые пользуются двумя продуктами банка, и теми, которые пользуются одним;

Нулевая гипотеза H0: Нет различий в доходах между клиентами, которые пользуются двумя продуктами банка, и теми, которые пользуются одним.  
Альтернативная гипотеза Н1: Есть различия в доходах между клиентами, которые пользуются двумя продуктами банка, и теми, которые пользуются одним.

Поделим клиентов по продуктам

```python
# Клиенты с одним продуктом
one_prod = clients.query('products==1')
# Клиенты с двумя продуктами
two_prod = clients.query('products==2')
# Клиенты с тремя продуктами
three_prod = clients.query('products==3')
```

Проверим распределение зарплат для наших клиентов.

```python
fig, axes = plt.subplots(1, 3, figsize=(10, 5))
sns.boxplot(data=one_prod, y='est_salary', ax=axes[0]);
sns.boxplot(data=two_prod, y='est_salary', ax=axes[1]);
sns.boxplot(data=three_prod, y='est_salary',  ax=axes[2]);

axes[0].set_title('1 продукт');
axes[1].set_title('2 продукта');
axes[2].set_title('3 продукта');
```

Проверим данную гипотезу t-критерием Стьюдента. Предварительно избавимся от выбросов.

```python
# Нижняя граница выбросов у нас ограничится нулем, определим верхнюю
n = 0
upper_bound_list = []
for i in [one_prod, two_prod, three_prod]:

    n += 1
    q1, q3 = np.percentile(i['est_salary'], [25, 75])
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    upper_bound_list.append(upper_bound)
    display(
        f'Верхняя граница нормальных значений зарплаты для пользователей, имеющих {n} продуктов:', upper_bound)
    display(f"Количество пользователей в 'выбросе' имеющих {n} продуктов", len(
        i.query('est_salary > @upper_bound')))
```

Ограничим выборку клиентами без выбросов и проведем тест.

```python
alpha = 0.05
results = st.ttest_ind(
    one_prod.query('est_salary>@upper_bound_list[0]')['est_salary'],
    two_prod.query('est_salary>@upper_bound_list[1]')['est_salary']
)
display('p-значение:', results.pvalue)
if results.pvalue < alpha:
    display('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя'
    )
```

**Итого:**
Есть статистическая разница в уровне дохода между клиентами, которые используют один продукт и клиентами, которые используют два продукта.

## Проверим аналогичную гипотезу, но для клиентов с двумя и тремя продуктами

```python
alpha = 0.05
results = st.ttest_ind(
    three_prod.query('est_salary>@upper_bound_list[2]')['est_salary'],
    two_prod.query('est_salary>@upper_bound_list[1]')['est_salary']
)
display('p-значение:', results.pvalue)
if results.pvalue < alpha:
    display('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя'
    )
```

Статистической разницы в уровне дохода нет между клиентами, которые пользуются двумя продуктами и клиетами, которые используют три продукта.

## Проверим гипотезу различия возраста между теми клиентами, которые пользуются двумя продуктами банка, и теми, которые пользуются одним.

Сформулируем нулевую и альтернативную гипотезы:  
Нулевая гипотеза Н0: Нет статистической разницы в возрасте между клиетами банка, которыепользуются двумяпродуктами банка, и теми, которые пользуются одним.  
Альтернативная гипотеза Н1: Есть статистическая разница в возрасте между клиетами банка, которыепользуются двумяпродуктами банка, и теми, которые пользуются одним.  

Проверим распределение возраста наших пользователей

```python
fig, axes = plt.subplots(1, 3, figsize=(10, 5))
sns.boxplot(data=one_prod, y='age', ax=axes[0]);
sns.boxplot(data=two_prod, y='age', ax=axes[1]);
sns.boxplot(data=three_prod, y='age',  ax=axes[2]);

axes[0].set_title('1 продукт');
axes[1].set_title('2 продукта');
axes[2].set_title('3 продукта');
```

```python
# Нижняя граница выбросов у нас ограничится нулем, определим верхнюю
n = 0
upper_bound_list = []
two_prod = two_prod.dropna(subset=['age'])
for i in [one_prod, two_prod, three_prod]:

    n += 1
    q1, q3 = np.percentile(i['age'], [25, 75])
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    upper_bound_list.append(upper_bound)
    display(
        f'Верхняя граница нормальных значений возраста для пользователей, имеющих {n} продуктов:', upper_bound)
    display(f"Количество пользователей в 'выбросе' имеющих {n} продуктов", len(
        i.query('age > @upper_bound')))
```

Отбросим выбросы и проведем тест.

```python
upper_bound_list
```

```python
alpha = 0.05
results = st.ttest_ind(
    one_prod.query('age>@upper_bound_list[0]')['age'],
    two_prod.query('age>@upper_bound_list[1]')['age']
)
display('p-значение:', results.pvalue)
if results.pvalue < alpha:
    display('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя'
    )
```

Нет статистической разницы в возрасте между пользователями одного продукта и поьзователями двух продуктов.

## Проверим гипотезу о том, что есть статистическая разница в возрасте между отточными и неотточными клиентами.

```python
df_churn_0 = clients.query('churn==0')
df_churn_0 = df_churn_0.dropna(subset=['age'])
df_churn_1 = clients.query('churn==1')
df_churn_1 = df_churn_1.dropna(subset=['age'])
```

```python
upper_bound_list = []
for i in [df_churn_0,df_churn_1]:
    q1, q3 = np.percentile(i['age'], [25, 75])
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    upper_bound_list.append(upper_bound)
    
display('Нормальный максимальный возраст для отточных',upper_bound_list[0])
display('Нормальный максимальный возраст для неотточных',upper_bound_list[1])

display(f"Количество пользователей в 'выбросе' отточных ", len(
        i.query('age > @upper_bound_list[0]')))
display(f"Количество пользователей в 'выбросе' неотточных", len(
        i.query('age > @upper_bound_list[1]')))
```

Сформулируем гипотезы:  
Нулевая гипотеза Н0: Нет статистической разницы между возрастом отточных и неотточных клиентов.  
Альтернативная гипотеза Н1: Статистическая разница между возрастом отточных и неотточных клиентов есть.

```python
alpha = 0.05
results = st.ttest_ind(df_churn_0['age'], df_churn_1['age'])
display('p-значение:', results.pvalue)

if results.pvalue < alpha:
    display('Отвергаем нулевую гипотезу')
else:
    display('Не получилось отвергнуть нулевую гипотезу')
```

Есть статистическая разница в возрасте отточных и неотточных клиентов.

## Проверим гипотезу о том, что есть статистическая разница в Score между отточными и неотточными клиентами.

Нулевая гипотеза Н0: Среденее значение Score у отточных и неотточных клиентов не отличается.  
Пльтернативная гипотеза Н1: Среденее значение Score у отточных и неотточных клиентов отличается. 

```python
alpha = 0.05
results = st.ttest_ind(df_churn_0['score'], df_churn_1['score'])
display('p-значение:', results.pvalue)

if results.pvalue < alpha:
    display('Отвергаем нулевую гипотезу')
else:
    display('Не получилось отвергнуть нулевую гипотезу')
```

Есть статистическая разница между средним значением кредитного скоринга между отточными и неотточными клиентами.

С точки зрения возвращения клиентов приоритетным для банка должны быть отточные клиенты.  
Для получения дополнительной прибыли за счет обората средств клиентов, банку должны быть интересны VIP-пользователи.  
Неотточные клиенты могут быть полезны банку для выявления портрета и целевой аудитории своих продуктов.

Первый сегмент - идеальный портрет неотточного клиента, может служить как ориентир целевой группы самых "верных" клиентов.  

Третий сегмент - VIP-клиенты. Интересны с позиции выгоды для банка.  

Если приоритизировать оставшиеся сегменты по уровню оттока:  

1. Второй сегмент: уровень оттока - 53%. Состоит из мужчин с 3-мя и более продуктами и оценкой собственности выше 3 ед (по 10-ти бальной шкале). Приоритизировать их нужно, потому что данные клиенты обладают самыми простыми и понятными характеристиками.  


2. Четвертый сегмент: уровень оттока - 42%. Здесь у нас мужчины, но уже с небольшим количеством продуктов (до 3-х), хорошим кредитным скорингом (860-900) и оценкой собственности (5 ед. из 10). Даже не смотря на то, что продуктов мало, данная группа показывает очень высокий уровень оттока, обладая очень хорошим кредитным скорингом и оценкой собственности выше средней.  


3. Пятый сегмент: уровень оттока - 40%. Это жители Ярославля, обладающие средней зарплатой 50 000 - 150 000 рублей (при медианной для города: 128 000 рублей) и балансом свыше 750 000 рублей (при медианном значении длявсех клиентов банка в 520 000 рублей.), но при этом не имеющие кредитной карты. Как показывают данные, отсутсвие кредитной карты у клиента - выше вероятность его ухода.  


4. Шестой сегмент: уровень оттока - 38%. Это клиенты предпенсионного возраста, имеющие хороший кредитный рейтинг, но малое число продуктов (3 или меньше).





# Рекомендации для каждого сегмента

## Первый сегмент

Пользователи первого сегмента, который условно можно назвать неотточным, имеет малое количество продуктов в пользовании, чаще всего имеет кредитную карту и по большей степени состоит из женщин.  

Можно для данного сегмента дать следующие рекомендации:  
- предоставить данным пользователям, особенно с учетом пола, дополнительную информацию о существующих продуктах банка, их преимуществах;  
- возможно рассмотрение создания нового продукта для данной целевой аудитории;  
- так как данные пользователи имеют высокий уровень кредитного скоринга и кредитные карты, но оценка собственности данных пользоватлей низка, то можно предлагать данным пользователям больше кредитных продуктов, чтобы увеличить количество собственности;  


## Второй сегмент

Пользователи второго (отточного) сегмента, на самом деле очень привлекательны для банка, так как в среднем имеют больше трех продуктов, высокую оценку собственности, относительно большой баланс. Рекомендации, в первую очередь, должны быть направлены на удержание таких клиентов.  
Для таких клиентов можно сформулировать следующие рекомендации:  
- предлагать данным пользователям оформить кредитную карту, так как продуктов у этих пользователей уже много, но доля наличия кредитных карт ниже, чем у пользователей банка в среднем;  
- так как у данных пользователей высокая оценка собственности, то можно рассмотреть возможность предложения им дополнительных страховых продуктов для имущества;

## Третий сегмент

Третий сегмент пользователей или VIP-пользователи особенно интересны банку из-за наличия большого количества средств на счетах. Для них характерно использование 2-3 продуктов, высокая оценка собственности и высокая зарплата. Этих пользователй можно назвать самымии лояльными, потому что среди нет отточных и неактивных клиентов.  
Банк может проводить мероприятия, которые укрепят связь с такими клиентами.  

Некоторые из мероприятий могут включать в себя:  
- эксклюзивные мероприятия и ивенты;
- особые услуги, такие как личные финансовые консультации, персональные менеджеры, эксклюзивные банковский продукты;  
- предложение льготных условий кредитования, льготные ставки по вкладам.

## Четвертый сегмент

Мужчины с малым числом продуктов, высоким кредитным рейтингом, высокой, относительно среднего, оценкой собственности. Уровень оттока - более 42%.  

Для таких клиентов банк может провести следующие мероприятия с целью уменьшит отток:  
- организовать мероприятия для мужчин. Спортивные события, концерты, мастер-классы;  
- предоставить привелегии для владельцев собственности. С учетом высокого кредитного рейтинга, можно предложить скидки на кредитные продукты, повышенные лимиты на кредитную карту и прочее. Это может повысить заинтересованность клиентов банком, а банку повысить количество продуктов у клиента;  
- предоставить консультационную поддержку в сфере управления личными финансами и инвестирования, так как данные клиенты имеют высокую оценку собственности.  

## Пятый сегмент

Ярославцы, с хорошей зарплатой, большим балансом, но без кредитки. Уровень оттока: 40%.
Примеры мероприятий для удержания такого рода клиентов:  
- клиенты, которые хранят большой баланс на своих счетах, могут быть заинтересованы в различных вкладах, облигациях и других инвестиционных продуктах, которые могут принести им дополнительный доход;  
- предоставление кредитной карты на привлекательных условиях, например с повышенным лимитом, который соответсвует большому балансу на счете;  
- проведение социальных мероприятий в Ярославле. Этот город лидируют по доле отточных клиентов. Мероприятия, направленные на заинтересованность клиента в банке или проведение опросов удовлетворенности в данном городе, помогут "задержать" ярославцев в "Метанпромбанке".

## Шестой сегмент

Клиенты предпенсионного возраста, с хорошим показателем кредитного рейтинга и небольшим числом продуктов (до 3-х включительно). Уровень оттока - 38%.  
Для данных клиетов с учетом их возраста, можно предложить:  
- персонализированного менеджера, который бы консультировал и подсказывал выбор оптимальных продуктов, объяснял бы сложности в работе с приложением и тд.;  
- предоставление персональных услуг в части управления бюджетом, составлении наследства;  
- предоставление возможности и обучение пользованию дистанционным онлайн обслуживанием.
